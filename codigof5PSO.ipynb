```python
import numpy as np

def demo_func(x):

    for i in range(30):
        x[i]

    soma = 0
    for i in range(30):
        soma += x[i]**2 - 10 * np.cos(2 * np.pi * x[i]) + 10
    return soma
```

```python
from sko.PSO import PSO

pso = PSO(func=demo_func, n_dim=30, pop=40, max_iter=150, lb=[-5.12] * 30, ub=[5.12] * 30, w=0.8, c1=0.5, c2=0.5)
pso.run()
print('best_x is ', pso.gbest_x, 'best_y is', pso.gbest_y)
```

```python
import matplotlib.pyplot as plt

plt.plot(pso.gbest_y_hist)
plt.show()
```
